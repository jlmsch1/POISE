# POISE v1.6 COSY fragment explorer

```
import sympy as sp
from pof_analyzer import Iz, pulse, evolve_J, classify_cosy_precursors, diagonal_precursors, cross_precursors, observable

tau, J = sp.symbols("tau J", real=True)
```

## 1. Build a simple COSY opening fragment

`Iz -> 90x(I) -> J(tau) -> 90x(I)`

```
expr = Iz()
expr = pulse(expr, angle=90, phase="x", spins="I")
expr = evolve_J(expr, time=tau)
expr = pulse(expr, angle=90, phase="x", spins="I")

print("Full expression:")
print(expr)
print()
print("Compact trig:")
print(expr.to_string(compact_trig=True))
print()
print("LaTeX compact:")
print(expr.to_latex(compact_trig=True))
```

## 2. Simple COSY-style precursor filtering

```
classes = classify_cosy_precursors(expr)
print("Diagonal precursors:")
print(classes["diagonal"])
print()
print("Cross precursors:")
print(classes["cross"])
print()
print("Other:")
print(classes["other"])
```

## 3. Observable filtering

```
print("Observable on I:")
print(observable(expr, detect_on="I"))
print()
print("Observable on S:")
print(observable(expr, detect_on="S"))
```

## 4. Another fragment: create mixed I/S terms, then classify

```
expr2 = Iz()
expr2 = pulse(expr2, angle=90, phase="x", spins="I")
expr2 = evolve_J(expr2, time=tau)
print(expr2)
print()
classes2 = classify_cosy_precursors(expr2)
print("Diagonal:", classes2["diagonal"])
print("Cross:", classes2["cross"])
print("Other:", classes2["other"])
```
